# To implement CART Algorithm in Python

```python
import numpy as np
import pandas as pd
from chefboost import Chefboost as chef
```

```python
df=pd.read_csv('datasets/mammals.csv')
df
```

```python
df.isnull().sum()
```

```python
#preprocessing
df.describe()
```

```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy='mean')
df['non_dreaming']=imp.fit_transform(pd.DataFrame(df['non_dreaming']))
df['dreaming']=imp.fit_transform(pd.DataFrame(df['dreaming']))
df['total_sleep']=imp.fit_transform(pd.DataFrame(df['total_sleep']))
df['life_span']=imp.fit_transform(pd.DataFrame(df['life_span']))
df['gestation']=imp.fit_transform(pd.DataFrame(df['gestation']))
```

```python
df.isnull().sum()
```

```python
#model building
config={'algorithm':'CART'}
model=chef.fit(df,config=config,target_label='danger')
```

```python
#prediction
pred=chef.predict(model,param=['Asianelephant',2547.000,4603.0,2.1,1.9,4.9,72.0,626.3,3,5])
pred
```

```python
pr2=chef.predict(model,param=['Vervet',4.190,58.0,9.7,0.6,10.3,24.0,210.0,4,3])
pr2
```
